```python
#import stable_baselines3 as sb
import os

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import tensorboard as tb
import pickle as pkl
sns.set()

```

```python
filenames = ['../outputs/resultats/csv/csv/cem_cartpole_merged',
            '../outputs/resultats/csv/csv/cem_erl_cartpole_merged',
            '../outputs/resultats/csv/csv/cem_rl_cartpole_merged',
            '../outputs/resultats/csv/csv/td3_cartpole_merged']
names = ["cem","cem_erl", "cem_rl","td3"]
```

```python
for filename, name in zip(filenames,names):
    data = pd.read_csv(filename, sep=",")
    data = pd.read_csv(filename, sep=",", names=list(range(data.shape[1])))
    x1 = list(data.groupby([0]).quantile(0.75)[1])
    x2 = list(data.groupby([0]).quantile(0.25)[1])
    x_mean = list(data.groupby([0]).mean()[1])
    x_std = list(data.groupby([0]).std()[1])
    plt.plot(np.array(range(len(x_mean))) * 5000,x_mean, label=name)
    plt.fill_between(np.array(range(len(x1)))* 5000, x1, x2, alpha=0.25)

plt.xlabel("step")
plt.ylabel("reward")
plt.legend(loc='lower right')
plt.title("Cartpole - Mean")
plt.savefig("Cartpole_Mean", format='pdf')
```
